### 2.2 Data Imputation: year_completed and max_floor_lvl
Some of our blocks have null values for  have these features: 'year_completed', 'max_floor_lvl', '1room_sold', '2room_sold','3room_sold', '4room_sold', '5room_sold', 'exec_sold','total_dwelling_units'.

In order to do a systematic inference of these values. We need to first impute the first two columns.

```python
# << IMPUTATION ATTEMPT 1: USING RESALE_PRICES DATASET>>
# << CREATING A NESTED FOR LOOP TO FILL IN FILL IN THE NaNs IN demo_all_updated>>
# << SPECIFICALLY: FOR year_completed AND max_floor_lvl... >>
# IDENTIFY OUR COLS WITH NaNs
cols_with_nans = demo_all_updated[demo_all_updated.total_dwelling_units.isnull()].index.tolist()
# CREATING EMPTY LISTS FOR OUR LOOPS TO APPEND INTO
year_list = []
lvl_list = []
# SORTING RESALE_PRICES BY storey_range
# THIS HELPS US GET TO THE HIGHER FLOORS FIRST
resale_prices = resale_prices.sort_values("storey_range", ascending=False)
num_of_blocks = demo_all_updated.shape[0]
# LOOPING THROUGH THE BLOCKS IN demo_all_updated
for row in range(num_of_blocks):
    match_score = 0
    # FOR COLUMNS WITH NaNs
    if row in cols_with_nans and match_score==0:
        demo_streets = demo_all_updated.street.iloc[row].split(" / ")
        demo_blk_no = demo_all_updated.blk_no.iloc[row]
        # LOOPING THROUGH resale_prices
        for resale_row in range(resale_prices.shape[0]):
            resale_street = resale_prices.street_name.iloc[resale_row]
            resale_block = resale_prices.block.iloc[resale_row]
            if match_score==0:
                # <<FOR A MATCHED BLOCK>>
                if (demo_blk_no == resale_block) and (resale_street in demo_streets):
                    print("<<MATCHED: ", demo_blk_no, demo_streets, ">>")
                    print("HEIGHT :", max(resale_prices.storey_range.iloc[resale_row].split(" to ")), " floors")
                    print("YEAR COMPLETED :",resale_prices.lease_commence_date.iloc[resale_row], "\n")
                    # APPEND lease_commence_date (AKA year_completed)
                    year_list.append(resale_prices.lease_commence_date.iloc[resale_row])
                    # APPEND THE HIGHEST FLOOR IN storey_range (AKA max_floor_lvl)
                    lvl_list.append(max(resale_prices.storey_range.iloc[resale_row].split(" to ")))
                    match_score+=1
            else:
                break
        # <<FOR BLOCKS WITH NO MATCHES>>
        if match_score == 0:
            print("NO MATCHES FOR: ", demo_blk_no, demo_streets)
            # APPEND ERROR MSGs FOR NON-MATCHES
            year_list.append("NO MATCHES")
            lvl_list.append("NO MATCHES")        
    # <<FOR CLEANED BLOCKS>> 
    else: 
        # SIMPLY APPEND EXISTING VALUES
        year_list.append(demo_all_updated["year_completed"].iloc[row])
        lvl_list.append(demo_all_updated["max_floor_lvl"].iloc[row])

# PRINT A REPORT        
print("COMPLETED YEARS: ", len(year_list),"/", num_of_blocks)
print("NUM OF YEARS WITH NO MATCHES: ", year_list.count("NO MATCHES"), "\n")

print("COMPLETED LEVELS: ", len(lvl_list),"/", num_of_blocks)
print("NUM OF LEVELS WITH NO MATCHES: ", lvl_list.count("NO MATCHES"), "\n")
```

```python
# UPDATING BOTH COLUMNS WITH NEW VALUES
demo_all_updated["year_completed"]=year_list
demo_all_updated["max_floor_lvl"]=lvl_list
```

```python
# << IMPUTATION ATTEMPT 2: MATCHING TO BLOCKS BUILT IN SAME ERA>>
year_list = []
lvl_list = []
num_matchless = 0
for row in range(demo_all_updated.shape[0]):
    if demo_all_updated.year_completed.iloc[row]=="NO MATCHES":
        match_score = 0
        demo_streets = demo_all_updated.street.iloc[row]
        demo_blk_no = demo_all_updated.blk_no.iloc[row]
        for compare in range(demo_all_updated.shape[0]):
            if demo_all_updated.year_completed.iloc[compare]!="NO MATCHES":
                comp_streets = demo_all_updated.street.iloc[compare]
                
                if demo_streets == comp_streets and match_score == 0:
                    comp_year = demo_all_updated.year_completed.iloc[compare] 
                    comp_lvl = demo_all_updated.max_floor_lvl.iloc[compare] 
                    year_list.append(comp_year)
                    lvl_list.append(comp_lvl)
                    match_score +=1
        if match_score == 0:
            print("STILL NO MATCHES FOR: ", demo_blk_no,demo_streets)
            year_list.append("NO MATCHES")
            lvl_list.append("NO MATCHES")
            num_matchless+=1
    else:
        year_list.append(demo_all_updated.year_completed.iloc[row])
        lvl_list.append(demo_all_updated.max_floor_lvl.iloc[row])
print("\nTOTAL WITHOUT MATCHES: ",num_matchless )
```

### 2.4 Data Imputation II : types of apartments
We will use a the district-wide mean of the following features for imputation: '1room_sold', '2room_sold','3room_sold', '4room_sold', '5room_sold', 'exec_sold','total_dwelling_units'.


```python
# <<APPEND MEAN DETAILS FROM ENTIRE DISTRICT>>

# IDENTIFY OUR COLS WITH NaNs
cols_with_nans = demo_all_updated[demo_all_updated.total_dwelling_units.isnull()].index.tolist()
df_list = []
total_dwell_list = []

# THESE ARE THE COLS THAT NEED INFORMATION
select_cols = ['1room_sold', '2room_sold','3room_sold', 
               '4room_sold', '5room_sold', 'exec_sold',
               ]
# CREATING A SCORE FOR ERRORS
odd_match =0
for row in range(demo_all_updated.shape[0]):
    if row in cols_with_nans and match_score==0:
        # DEFINING CONDITIONS
        demo_blk = demo_all_updated["blk_no"].iloc[row]
        demo_street = demo_all_updated["street"].iloc[row]
        d_lvl = int(demo_all_updated["max_floor_lvl"].iloc[row])
        d_year = int(demo_all_updated["year_completed"].iloc[row])
        demo_town = demo_all_updated["bldg_contract_town"].iloc[row]
        # DEFINING MASKS
        town_mask = (stand_blocks.bldg_contract_town == demo_town)
        height_mask = (stand_blocks.max_floor_lvl.isin(range(d_lvl-2, d_lvl+2)))
        year_mask = (stand_blocks.year_completed.isin(range(d_year-2, d_year+2)))
        # WE'LL GET THE MEAN OF BLOCKS...
        # ...BUILT WITHIN THE SAME PERIOD(+- 2 YEARS) AND OF SIMILAR HEIGHT(+- 2 FLOORS)
        mean_dwell = round(stand_blocks[height_mask&year_mask].describe().loc["mean"][select_cols])

        
        # <<CRITERIA: APPEND MEAN DETAILS FROM BLOCKS WITHIN TOWN IF POSSIBLE>>
        if (sum(mean_dwell) != 0.0) or (len(df_list)!=0) or math.isnan(sum(mean_dwell))==False:
            total_dwell_list.append(sum(mean_dwell))
            df_list.append(mean_dwell)
            print("<<CLOSE MATCH>>")
            print("<<", demo_blk, demo_street, ">>")
            print("Year Completed: ",demo_all_updated["year_completed"].iloc[row] )
            print("Height: ",demo_all_updated["max_floor_lvl"].iloc[row] )
            print("Total Dwelling Units: ",sum(mean_dwell) )
            print(mean_dwell, "\n")
        # APPEND ERROR VALUE FOR ERRORS
        else:
            print("<<ERROR!>>")
            print("<<", demo_blk, demo_street, ">>\n")
            df_list.append(demo_all_updated.loc[row][select_cols])
            total_dwell_list.append(demo_all_updated.total_dwelling_units.iloc[row])
            odd_match+=1
    # APPEND CURRENT VALUE IF INFORMATION EXISTS
    else:
        df_list.append(demo_all_updated.loc[row][select_cols])
        total_dwell_list.append(demo_all_updated.total_dwelling_units.iloc[row])
# CREATING A DATAFRAME WITH ALL THE FEATURES APPENDED
df= pd.DataFrame(df_list)
df["total_dwelling_units"]=total_dwell_list
df
# PRINT COMPLETION REPORT
print("COMPLETED: ", df.shape[0], " / ", demo_all_updated.shape[0])
print("Odd Matches: ", odd_match)
```

### 2.5 Accounting for discrepancies in total_dwelling_units
The sum of our flat types are supposed to add up to the total dwelling units. Some of our blocks might have rental units that don't appear in the flat-types column. Let's create a column that accounts for that. 

```python
def create_other_sold_col(df, new_other_list):
    for row in range(df.shape[0]):
        demo_types_sum = df[['1room_sold', '2room_sold', 
                                         '3room_sold', '4room_sold', 
                                         '5room_sold', 'exec_sold']].iloc[row].sum()
        total_dwell = df.total_dwelling_units.iloc[row]

        if demo_types_sum != total_dwell:
            print("DOES NOT COMPUTE!")
            print(df[["blk_no","street"]].iloc[row])
            print("DIFFERENCE: ", total_dwell-demo_types_sum, "\n")
            new_other_list.append(total_dwell-demo_types_sum)
        else:
            new_other_list.append(0)
    return new_other_list
```

```python
demo_all_clean["other_sold"]=0
new_other_list = []

demo_all_clean["other_sold"] = create_other_sold_col(demo_all_clean, new_other_list)
```

## 3 Balancing both datasets 
Our next task involves making sure both our standing blocks and demolished blocks datasets have the same number of features.

### 3.1 Creating a column for other_sold
The stand_blocks data set has a lot more information about rental units that our demolished blocks do not have. Let's capture that in an "other_sold" column. 

```python
# CREATING AN other_sold COLUMN
stand_blocks["other_sold"]=0
new_other_list = []

stand_blocks["other_sold"] = create_other_sold_col(stand_blocks, new_other_list)
```

### 3.2 Removing demolished blocks from the standing blocks dataset

Our standing blocks dataset contains a fair number of blocks that have already been slated for demolition. Let's remove them to avoid any kind of conflict. 

```python
# <<CREATING A FOR LOOP TO FISH OUR DOUBLE-MENTIONED BLOCKS>>
rows_to_drop = []
for row in range(stand_blocks.shape[0]):
    # DEFINING CONDITIONS FOR STAND BLOCKS
    stand_street = stand_blocks.street.iloc[row]
    stand_blk = stand_blocks.blk_no.iloc[row]
    stand_year = stand_blocks.year_completed.iloc[row]
    for demo in range(demo_all_clean.shape[0]):
        # DEFINING CONDITIONS FOR DEMO BLOCKS
        demo_street = demo_all_clean.street.iloc[demo].split(" / ")
        demo_blk = demo_all_clean.blk_no.iloc[demo]
        # CREATING A RANGE OF YEARS FOR SOME LEEWAY FOR ERRORS
        demo_year = int(demo_all_clean.year_completed.iloc[demo])
        demo_year_range = range(demo_year-2, demo_year+2) 
        # ISOLATING DOUBLE-MENTIONED BLOCKS
        if (stand_blk == demo_blk) and (stand_street in demo_street) and (stand_year in demo_year_range):
            # APPEND ROW_NUM INTO A LIST
            rows_to_drop.append(stand_blocks.index[row])
            print("<<DOUBLE-MENTIONED BLOCKS>>")
            print("DEMO BLOCK: ", demo_blk, demo_street)
            print("DEMO BLOCK: ", stand_blk, stand_street)
            print("<<PLS REMOVE FROM STAND BLOCKS DATASET>>\n")
            
print("-"*15)
print("TOTAL DOUBLE-MENTIONED BLOCKS: ", len(rows_to_drop))
```

```python
# REMOVING DOUBLE-MENTIONED BLOCKS
print("SHAPE BEFORE", stand_blocks.shape)
stand_blocks_clean = stand_blocks.drop(rows_to_drop, axis="rows")
print("SHAPE AFTER", stand_blocks_clean.shape)
# WE SHOULD SEE THAT 43 BLOCKS HAVE BEEN DROPPED
```

### 3.4 Creating latest_survival_date for standing blocks
